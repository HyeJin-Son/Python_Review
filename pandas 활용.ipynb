```python
#from emp join dept dept using(deptno)
emp = pd.read_csv('C:\ds_work\data\emp.csv')
dept = pd.read_csv('C:\ds_work\data\dept.csv') 
result = pd.merge(emp,dept,on='deptno')
#group by Loc; select loc, sum(sal)

result02=result.groupby('loc')['sal'].sum().reset_index()
print(result02)
```

```python
print('-----------inner join-------------')
result1=pd.merge(emp,dept, on='deptno')
print(result1)
print('-----------outer join-------------')
result2=pd.merge(emp,dept,how='outer',on='deptno')
print(result2)
print('-----------Right join-------------')
result =pd.merge(emp,dept, how='right', on='deptno')
print(result)
result02=result.groupby('loc')['sal'].sum().reset_index()
print(result02)
print('-----------Left join-------------')
result3 =pd.merge(emp,dept, how='left', on='deptno')
print(result3)
result03=result.groupby('loc')['sal'].sum().reset_index()
print(result03)
```
