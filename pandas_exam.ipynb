```python
#3.직업이 'SALESMAN'인 사원의 이름과 직업, 봉급을 출력해보자.
#select ename, job, sal from emp where job ='SALEMAN';

emp[['ename','job','sal']][emp['job']=='SALESMAN']

```

```python
#4. 월급이 3000이상인 사원의 이름과 봉급을 출력하자.
#select ename, sal from emp where sal >=3000;

emp[['ename','sal']][emp['sal']>=3000]
```

```python
#5. 사원의 이름과 월급을 출력하되 1000 이상에서 3000 까지 출력하자
#select ename, sal from emp where sal >=1000 and sal <= 3000;
#select ename, sal from emp where sal between 1000 and 3000;
emp[['ename','sal']][emp['sal'].between(1000,3000)]
```

```python
#6. 사원의 이름과 월급을 출력하되 1000 이상에서 3000 이 아닌 목록을 출력하자.
#select ename, sal from emp where sal <1000 and sal > 3000;
#select ename, sal from emp where sal not between 1000 and 3000;
emp[['ename','sal']][~emp['sal'].between(1000,3000)]
```

```python
#8) 커미션이 책정되지 않은 사원의 이름과 봉급 커미션을 출력하자.

emp[['ename','sal','comm']] [emp['comm'].isnull()]
```

```python
#9) 사원의 이름이 'SCOTT'인 사원의 이름과 봉급을 출력

emp[['ename','sal']][emp['ename']=='SCOTT']
```

```python
#10. 이름만 추출해서 첫글자만 출력해보자.

for i in emp['ename']:
    print(i, i[0] , i[-1])
```

```python
#10. 이름만 추출해서 마지막 글자가 T로 끝나는 이름 출력해보자.

for i in emp['ename']:
    if i[-1]=='T':
        print(i)

```
