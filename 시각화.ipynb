```python
#seaborn 
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#dropna : 값이 없는 행/열을 전체 삭제
titanic['deck'].value_counts(dropna=False)
```

```python
titanic.isnull()#결측치 확인
```

```python
#누락데이터 개수 isnull().sum()
titanic.isnull().sum()
```

```python
#누락데이터 행, 열 삭제 dropna(axis, thresh=누락데이터 개수) -> 데이터 과적합
# 대상 ->시뮬레이션 99.8% 정확도(테스트) -> 서비스 배포
titanic = titanic.dropna(axis=1, thresh=500)
titanic.head()
#deck이 사라짐을 확인
```

```python
#결측(누락데이터) 치환 df.fillna(대체값, inplace=True), 최대값, 최소값, 평균값등등 대처
age_mean=titanic['age'].mean()
age_mean
```

```python
titanic['age'].fillna(age_mean, inplace=True)
```

```python
titanic.isnull().sum()
```

```python
#결측(누락데이터): 최빈값으로 구해서 치환
titanic['embarked'].value_counts(dropna=False)
```

```python
titanic['embarked'].value_counts(dropna=False).idxmax()
```

```python
titanic['embarked'].value_counts(dropna=False).idxmin()
```

```python
e_idxmax = titanic['embarked'].value_counts(dropna=False).idxmax()
titanic['embarked'].fillna(e_idxmax,inplace=True)
```

```python
#embark_town 도 최빈값으로 채워보자.
e_idxmax = titanic['embark_town'].value_counts(dropna=False).idxmax()
titanic['embark_town'].fillna(e_idxmax,inplace=True)
```
