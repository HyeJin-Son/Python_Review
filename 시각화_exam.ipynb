```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
emp = pd.read_csv('C:\ds_work\data\emp.csv')
emp
```

```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # '맑은 고딕'으로 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = emp['ename']
y = emp['sal']

plt.plot(x, y) 
plt.grid(True)
#X축 LABEL의 기울기
plt.xticks(rotation=45)
plt.xlabel('사원명') 
plt.ylabel('봉급') 
plt.title('사원별 봉급')
plt.show()

```

```python
result = emp.groupby('job')['sal'].mean().reset_index()
print(result)
explode_value = (0.1, 0, 0, 0, 0) 

plt.figure(figsize=(5,5)) # 그래프의 크기를 지정
plt.pie(result['sal'], labels=result['job'], autopct='%.1f%%', startangle=90, counterclock = False, explode=explode_value, shadow=True)
plt.show()

```

```python
rst=emp.groupby('deptno')['sal'].sum().reset_index()
#print(rst)


plt.bar(rst['deptno'],rst['sal'],width=5,tick_label=rst['deptno'])
plt.grid(True)
plt.xlabel("부서번호") 
plt.ylabel("봉급의 합") 
plt.title("부서별 봉급의 합 막대 그래프") 
plt.show()

```

```python
#직업별 봉급의 최대값
job_sal_max=emp.groupby('job')['sal'].max().reset_index()
#print(job_sal_max)
#부서별 봉급의 최대값
dept_sal_max=emp.groupby('deptno')['sal'].max().reset_index()
#print(dept_sal_max)
#print(dept_sal_max.info())

plt.plot(job_sal_max['job'],job_sal_max['sal'],linestyle='--',color='blue',marker='o',label='직업별')
plt.legend()
plt.title('직업별 봉급의 최대값')
plt.xlabel('직업')
plt.ylabel('최대 봉급')
plt.grid(True)
# 그래프 분리
plt.figure()

dsm=dept_sal_max['deptno'].astype('str')
plt.plot(dsm,dept_sal_max['sal'],linestyle=':',color='green',marker='*',label='부서별')
plt.legend()
plt.title('부서별 봉급의 최대값')
plt.xlabel('부서')
plt.ylabel('최대 봉급')
plt.grid(True)
```
